```python
# Lista de 11 compras
my_basket = [['bread', 'butter', 'wine', 'bananas', 'coffee', 'carrots'],
             ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
             ['beer', 'chips', 'asparagus', 'salsa', 'milk', 'apples'],
             ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
             ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'orange juice'],
             ['bread', 'butter', 'beer', 'chips', 'milk'],
             ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
             ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
             ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
             ['butter', 'beer', 'chips', 'asparagus', 'apples'],
             ['tomatoes', 'onion', 'beer', 'chips', 'milk', 'coffee']]
```

```python
# Ciclo para obtener los productos únicos
productos_unicos = []
for i in range(len(my_basket)):
    for j in range(len(my_basket[i])):
        if my_basket[i][j] not in productos_unicos:
            productos_unicos.append(my_basket[i][j])

# Ciclo para obtener nombres de compras
index = []
for i in range(len(my_basket)):
    index.append("Compra No." + str(i+1))

# Creación de dataframe con ceros
df_my_basket = pd.DataFrame(columns = productos_unicos, index=index).fillna(0)

# Se rellena con 1 la información que corresponde
for i in range(len(my_basket)):
    for j in my_basket[i]:
        df_my_basket[j][i] = 1
df_my_basket
```

```python
# Aplicación del algoritmo "a priori" para localizar los artículos frecuentemente comprados en la base de datos
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Análisis "a priori" con un soporte mínimo del 3%
articulos_frecuentes = apriori(df_my_basket, min_support=0.03,
                              use_colnames = True).sort_values("support", ascending=False).reset_index(drop=True)
articulos_frecuentes["length"] = articulos_frecuentes["itemsets"].apply(lambda x:len(x))
articulos_frecuentes
```

```python
# Obtención de patrones entre productos
association_rules(articulos_frecuentes, metric = "confidence",
                 min_threshold = 0.9).sort_values("confidence", ascending = False).reset_index(drop=True)
```
